```python
#交通银行理财产品爬虫
import requests
import pandas
from bs4 import BeautifulSoup
import time

url = 'http://www.bankcomm.com/BankCommSite/jyjr/cn/lcpd/queryFundInfoListNew.do'

shuju = []

r = requests.get(url)

r.encoding = r.apparent_encoding

#print(r.text)

soup = BeautifulSoup(r.text, 'html.parser')

#print(soup.prettify()) 

def get_info():
    
    for i in range(0, 106):
        
        info = {}
    
        title0 = soup.find_all('h1', 'ellipsis')[i]
    
        title1 = str(title0)
    
        soup1 = BeautifulSoup(title1, 'html.parser')
    
        title2 = soup1.get_text()
    
        info['产品名称'] = title2
        
        money0 = soup.find_all('span','nb')[i*3]
    
        money1 = str(money0)
    
        soup2 = BeautifulSoup(money1, 'html.parser')

        money2 = soup2.get_text()
    
        info['起售金额'] = money2
        
        risk0 = soup.find_all('span','nb')[i*3+1]
    
        risk1 = str(risk0)
    
        soup3 = BeautifulSoup(risk1, 'html.parser')

        risk2 = soup3.get_text()
    
        info['风险等级'] = risk2
        
        rate0 = soup.find_all('span','nb')[i*3+2]
    
        rate1 = str(rate0)
    
        soup4 = BeautifulSoup(rate1, 'html.parser')

        rate2 = soup4.get_text()
        
        info['预期收益率'] = rate2
        
        cus0 = soup.find_all('div','customer_box')[i]
    
        cus1 = str(cus0)
    
        soup5 = BeautifulSoup(cus1, 'html.parser')

        cus2 = soup5.get_text()
    
        cus3 = cus2.strip()

        info['销售对象'] = cus3
        
        link0 = soup.find_all('a')[i]
    
        link1 = link0.get('href')
        
        info['详情链接'] = link1
        
        yield info
        

def write_to_excel(shuju):
    df = pandas.DataFrame(shuju)
    df.to_excel('交行产品爬虫v2.xlsx')  

    
def main():
    print("执行开始".center(150//2,'-'))
    start = time.perf_counter()
    mores = get_info()
    for more in mores:
        shuju.append(more)       
    write_to_excel(shuju)
    dur = time.perf_counter() - start
    print("数据爬取总耗时{:.2f}s".format(dur))
    print("\n"+"执行结束".center(150//2,'-'))

if __name__ == '__main__':
    main()
```
