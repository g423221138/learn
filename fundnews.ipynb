```python
#东方财富基金新闻爬虫

import requests
import re
from lxml import etree


def getinfo(url2):
    
    r = requests.get(url2)

    r.encoding = r.apparent_encoding

    html = etree.HTML(r.text)

    href = html.xpath('/html/body/div[6]/div[2]/div[1]/div/div[1]/div/ul/li/a/@href')

    for dizhi in href:
        newsurl = url0 + dizhi
        #print(newsurl)
        
        news = requests.get(newsurl)
        
        html = etree.HTML(news.text)
        
        title = html.xpath('/html/body/div[1]/div[4]/div[1]/div[2]/div[1]/h1/text()')
        
        time = html.xpath('/html/body/div[1]/div[4]/div[1]/div[2]/div[1]/div[1]/div[1]/div[1]/text()')
        
        content = html.xpath('//*[@id="ContentBody"]/p/text()')
        
        #print(type(content))
        
        biaoti = " ".join(title)
        
        shijian = " ".join(time)
        
        zhengwen = " ".join(content)
        
        #print(type(zhengwen))
        
        #print(zhengwen)
        
        pattern = r'中基协'    #单引号内输入需要检索的关键字
              
        match = re.search(pattern, zhengwen)
              
        if match:
            
            print('【标题】', biaoti)
            
            print('【时间】', shijian)
            
            print('【链接】', newsurl)
            
            print('【正文】', zhengwen)
            
    

url = 'http://fund.eastmoney.com/a/csmjj_{}.html'

url0 = 'http://fund.eastmoney.com/a/'
        
for p in range(1,2):     #括号内右侧数字可以修改，表示需要爬到第几页，数字1不变
    url2 = url.format(p)
    getinfo(url2)
```
