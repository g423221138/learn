```python
#用益信托网信息爬取

import requests
import json
from lxml import etree

url = 'http://www.yanglee.com/Action/ProductAJAX.ashx?mode=statistics&pageSize=40&pageIndex={}&conditionStr=producttype%3A1&start_released=&end_released=&orderStr=1&ascStr=ulup&_=1547452718827'

url1 = 'http://www.yanglee.com/Product/Detail.aspx?id='

headers = {
    'Accept': '*/*',
    'Accept-Encoding': 'gzip, deflate',
    'Accept-Language': 'zh-CN,zh;q=0.9',
    'Connection': 'keep-alive',
    'Cookie': 'safedog-flow-item=4F9739A28BFFB7B6774E420EF8B9BEAA; UM_distinctid=1684b5c101b205-0737359e6372ae-b781636-1fa400-1684b5c101c27; CNZZDATA1309589=cnzz_eid%3D1255970783-1547450167-http%253A%252F%252Fwww.yanglee.com%252F%26ntime%3D1547450167',
    'Host': 'www.yanglee.com',
    'Referer': 'http://www.yanglee.com/Product/Index.aspx',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98 Safari/537.36',
    'X-Requested-With': 'XMLHttpRequest'
}

for p in range(1,3):    #都是算头不算尾
    
    url0 = url.format(p)
    
    r = requests.get(url0, headers = headers)

    r.encoding = r.apparent_encoding

    d = r.json()

    #print(d)



    for n in range(0,40):     #一页40个产品
    
        #print(d.get('result')[n].get('ID'))
    
        url2 = url1 + d.get('result')[n].get('ID')
    
        #print(url2)
    
        r2 = requests.get(url2)
    
        r2.encoding = r2.apparent_encoding
    
        #print(r2.text)
    
        html = etree.HTML(r2.text)
    
        title = html.xpath('//*[@id="procon1"]/table/tr[1]/td[2]/text()')
    
        jigou = html.xpath('//*[@id="procon1"]/table/tr[3]/td[2]/text()')
    
        fenhong = html.xpath('//*[@id="procon1"]/table/tr[4]/td[4]//text()')
    
        time = html.xpath('//*[@id="procon1"]/table/tr[5]/td[2]/text()')
    
        qixian = html.xpath('//*[@id="procon1"]/table/tr[7]/td[2]/text()')
    
        rate = html.xpath('//*[@id="procon1"]/table/tr[8]/td[2]/text()')
    
        leixing = html.xpath('//*[@id="procon1"]/table/tr[9]/td[2]/text()')
    
        tuoguan = html.xpath('//*[@id="procon1"]/table/tr[9]/td[4]/text()')
    
        diqu = html.xpath('//*[@id="procon1"]/table/tr[10]/td[4]/text()')
    
        liangdian = html.xpath('//*[@id="procon1"]/table/tr[11]/td[2]/p/text()')
    
        others = html.xpath('//*[@id="procon1"]/table/tr[12]/td[2]/p/text()')
    
        title1 = " ".join(title)
    
        jigou1 = " ".join(jigou)
    
        fenhong1 = " ".join(fenhong)
    
        time1 = " ".join(time)
    
        qixian1 = " ".join(qixian)
    
        rate1 = " ".join(rate)
    
        leixing1 = " ".join(leixing)
    
        tuoguan1 = " ".join(tuoguan)
    
        diqu1 = " ".join(diqu)
    
        liangdian1 = " ".join(liangdian)
    
        others1 = " ".join(others)
    
        print('【产品名称】：',title1)
    
        print('【发行机构】：',jigou1)
    
        print('【收益分配】：',fenhong1)
    
        print('【发行时间】：',time1)
    
        print('【产品期限】：',qixian1)
    
        print('【预计收益】：',rate1)
    
        print('【投资方式】：',leixing1)
    
        print('【资金托管行】：',tuoguan1)
    
        print('【投资项目所在地】：',diqu1)
    
        print('【产品特点】：',liangdian1)
    
        print('【其他相关信息】：',others1)
    
        print(url2)
    
        print('-------------------------------------------------------------------')

    



```
