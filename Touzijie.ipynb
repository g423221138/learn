```python
#投资界信息爬取
import requests
import pandas
from lxml import etree
from bs4 import BeautifulSoup
import time
    
url = 'https://zdb.pedaily.cn/inv/p{}/'

shuju = []


def get_info(p):
    
    url1 = url.format(p)
    
    for i in range(1,25):
        
        try:
            
            r = requests.get(url1,timeout = 12)

            info = {}

            r.encoding = r.apparent_encoding

            html = etree.HTML(r.text)

            info['项目'] = " ".join(html.xpath(('//*[@id="inv-list"]/li[{}]/div[1]/a/text()').format(i)))

            info['投资方'] = " ".join(html.xpath(('//*[@id="inv-list"]/li[{}]/div[3]/a/text()').format(i)))

            info['阶段'] = " ".join(html.xpath(('//*[@id="inv-list"]/li[{}]/div[2]/span[1]/text()').format(i)))

            info['规模'] = " ".join(html.xpath(('//*[@id="inv-list"]/li[{}]/div[2]/span[2]/span[1]/text()').format(i)))

            info['地区'] = " ".join(html.xpath(('//*[@id="inv-list"]/li[{}]/div[2]/span[3]/text()').format(i)))
    
            info['融资时间'] = " ".join(html.xpath(('//*[@id="inv-list"]/li[{}]/div[4]/text()').format(i)))  
    
            link0 = 'https://zdb.pedaily.cn/'

            link1 = " ".join(html.xpath(('//*[@id="inv-list"]/li[{}]/div[1]/a/@href').format(i)))
    
            link2 = link0 + link1

            info['项目链接'] = link2
            
            r2 = requests.get(link2, timeout = 12)
            
            r2.encoding = r2.apparent_encoding
            
            soup = BeautifulSoup(r2.text, 'html.parser')

            a = soup.find_all('li')[14]

            b = str(a)

            c = b.lstrip('<li><span>所属行业：</span>').rstrip('</li>')
            
            info['所属行业'] = c
            
            print("已爬取第【{}】个项目信息，项目链接：".format(i), link2)
    
            yield info
        
        except Exception as e:
            
            print('获取信息出错', str(e))

            
def write_to_excel(shuju):
    df = pandas.DataFrame(shuju)
    df.to_excel('投资界爬虫5.xlsx')    

    
def main():
    print("执行开始".center(150//2,'-'))
    start = time.perf_counter()
    for p in range(1,21):
        print("正在爬取第【{}】页数据".format(p))
        mores = get_info(p)
        for more in mores:
            shuju.append(more)
        print("已完成第【{}】页数据爬取".format(p))
        print("----------------------------------------------------------------------------")
    write_to_excel(shuju)
    dur = time.perf_counter() - start
    print("数据爬取总耗时{:.2f}s".format(dur))
    print("\n"+"执行结束".center(150//2,'-'))

if __name__ == '__main__':
    main()



```
